```python
import os

# Replace 'images_folder' with the actual name or path of your folder
folder_path = '/content/drive/MyDrive/plant_pathology/images'
```

```python
# List all files in the folder
files = os.listdir(folder_path)
print("Files in folder:", files)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Check if there is at least one file in the folder
if files:
    # Full path to the first image file in the list
    image_path = os.path.join(folder_path, files[0])

    # Open the image using PIL
    img = Image.open(image_path)

    # Display the image using matplotlib
    plt.figure(figsize=(6,6))
    plt.imshow(img)
    plt.axis('off')  # Turn off axis numbers and ticks
    plt.show()
else:
    print("No files found in the folder.")
```

```python
# Filter and sort the list of image files
image_files = [f for f in files if f.endswith(('.png', '.jpg', '.jpeg'))]
image_files.sort()  # optional: sort by file name

# Loop over only the first 10 image files
for image_file in image_files[:10]:
    image_path = os.path.join(folder_path, image_file)
    img = Image.open(image_path)

    plt.figure(figsize=(4, 4))
    plt.imshow(img)
    plt.title(image_file)
    plt.axis('off')
    plt.show()
```
